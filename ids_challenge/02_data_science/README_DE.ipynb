### Datensatz

Den für dein Prognosemodell relevanten Datensatz findest du unter
`production_dataset/robot_energy_train.csv`.

Dieser enthält Daten zu durchgeführten Transportaufträgen und dem dabei gemessenen Energieverbrauch.
Ziel ist es, ein Machine-Learning-Modell zu entwickeln, das den Energieverbrauch anhand gegebener Auftragsmerkmale möglichst präzise vorhersagen kann.

Der Datensatz hat die folgenden Spalten:

| Spaltenname | Datentyp | Beschreibung |
|---|---|---|
| **Mission ID** | int | Eindeutiger Bezeichner für jeden Transportauftrag. |
| **Distance [m]** | float | Zurückzulegende Distanz in Metern. |
| **Elevation [m]** | float | Höhenunterschied zwischen Start- und Zielpunkt in Metern (negativ = bergab, positiv = bergauf). |
| **Cargo [kg]** | float | Gewicht der zusätzlichen Beladung des Roboters bei der Auftragsdurchführung in kg. |
| **Battery Level** | float | Batteriestand des Roboters zu Beginn des Auftrags im Wertebereich [0, 1]. Batteriestände unter 0,2 (20 %) können zu einem erhöhten Energieverbrauch führen. |
| **Time of Day [h]** | int | Stunde des Tages zum Auftragsstart (0–23). Kann relevant sein, da zwischen 9:00 und 17:00 Uhr erhöhter Betrieb der Logistikmitarbeiter zusätzliche Ausweichmanöver erfordern kann. |
| **Traffic Level** | str | Verkehrsaufkommen im Korridor zum Auftragszeitpunkt: `low`, `medium` oder `high`. |
| **Mode** | str | Betriebsmodus des Roboters: `normal` oder `priority` (Prioritätsmodus nutzt eine schnellere, aber weniger energieeffiziente Route). |
| ***Energy [kJ]*** | float | Beim Auftrag verbrauchte Energie in Kilojoule. **Vorhersageziel.** |

Der Datensatz enthält in einigen Feature-Spalten eine kleine Anzahl fehlender Werte.

**Modellierungshinweis:** In den Daten scheint es einen Bergauf-Last-Effekt zu geben, insbesondere wenn positive Steigung mit höherer Zuladung zusammenkommt.
Vielleicht ist dadurch in manchen Fällen ein einfaches Feature Engineering (z.B. ein Bergauf-Last-Feature oder ein binäres Feature, dass eine Uhrzeit zwischen 9:00 und 17:00 indiziert) sinnvoll.

Eine ungelabelte Version eines separaten Datensatzes steht unter `production_dataset/robot_energy_unlabeled.csv` bereit.
Diese Datei hat dieselben Spalten wie die Trainingsdaten, jedoch ohne die Zielspalte `Energy [kJ]`.
Nutze diese Datei, um mit deinem trainierten Modell Vorhersagen zu erzeugen, und reiche sie zusammen mit deinem Notebook ein.

### Anforderungen

Der Produktionsdatensatz enthält historische Daten zu von einem Roboter durchgeführten Aufträge.
Dein Ziel ist es, Regressionsmodelle zu trainieren und zu vergleichen, die den Energieverbrauch eines Transportauftrags anhand seiner Merkmale vorhersagen.

Es müssen mindestens **drei verschiedene Modelltypen** trainiert und verglichen werden, darunter mindestens ein parameterbasiertes Modell (z.B. lineare Regression) und ein baumbasiertes Modell (z.B. Decision Tree oder Random Forest).

Gib für alle trainierten Modelle **geeignete Fehlermetriken** an.
Achte darauf, dass diese Metriken nur dann aussagekräftig sind, wenn sie auf Daten berechnet werden, die nicht bereits im Modelltraining verwendet wurden.

**Entscheide dich für eines der Modelle, das du für den Produktiveinsatz vorschlagen möchtest** — und begründe deine Entscheidung.

Nutze abschließend dein gewähltes Modell, um Vorhersagen für den ungelabelten Datensatz `robot_energy_unlabeled.csv` zu erzeugen, und speichere diese in einer CSV-Datei.
Reiche diese Vorhersagedatei zusammen mit deinem Notebook ein.
