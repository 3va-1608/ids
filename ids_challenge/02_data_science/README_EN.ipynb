### Dataset

The data set relevant for your forecast model can be found at
`production_dataset/robot_energy_train.csv`.

This contains data on completed transport jobs and the energy consumption measured during each job.
The aim is to develop a machine learning model that can predict energy consumption as accurately as possible based on given job characteristics.

The data set has the following columns:

| Column name | Data type | Description |
|---|---|---|
| **Mission ID** | int | Unique identifier for each transport job. |
| **Distance [m]** | float | Distance to be covered in meters. |
| **Elevation [m]** | float | Height difference between start and destination point in meters (negative = downhill, positive = uphill). |
| **Cargo [kg]** | float | Weight of additional load carried by the robot during job execution in kg. |
| **Battery Level** | float | Battery charge level of the robot at the start of the job, in the range [0, 1]. Battery levels below 0.2 (20%) may result in increased energy consumption. |
| **Time of Day [h]** | int | Hour of the day at which the job started (0–23). May be relevant since between 9:00 and 17:00 increased activity by logistics employees can require additional evasive maneuvers. |
| **Traffic Level** | str | Level of corridor traffic at the time of the job: `low`, `medium`, or `high`. |
| **Mode** | str | Operating mode of the robot: `normal` or `priority` (priority mode uses faster but less energy-efficient routing). |
| ***Energy [kJ]*** | float | Energy consumed in kilojoules during job execution. **Prediction target.** |

The data set contains a small number of missing values in some feature columns.

**Modeling note:** The data seems to include stronger uphill strain effects, especially when uphill movement is combined with higher cargo.
Thus, feature engineering (for example and an uphill-strain feature or a binary high-demand-time indicator, which is positive for time of day values between 9:00 and 17:00) might allow for improved resutls.

An unlabeled version of a separate data set is provided in `production_dataset/robot_energy_unlabeled.csv`.
This file has the same columns as the training data, but without the `Energy [kJ]` target column.
Use this file to generate predictions with your trained model and submit them together with your notebook.

### Requirements

The production data set contains historic data regarding jobs performed by transport robot.
Your goal is to train and compare regression models that predict the energy consumption of a transport job based on its characteristics.

At least **three different model types** must be trained and compared, including at least one parameter-based model (e.g. linear regression) and one tree-based model (e.g. decision tree or random forest).

Specify **appropriate error metrics** for all trained models.
Make sure that such error metrics are only meaningful if they are calculated on data that has not already been used in model training.

**Decide on one of the models that you want to propose for production use** — justify your decision!

Finally, use your chosen model to generate predictions for the unlabeled data set `robot_energy_unlabeled.csv` and save the predictions to a CSV file.
Submit this prediction file together with your notebook.
