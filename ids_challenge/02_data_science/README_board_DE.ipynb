Um am Leaderboard teilzunehmen, müsst Ihr nur Eure Vorhersagen für den ungesehenen Datensatz auf unseren Server hochladen (unter Angabe von Gruppennamen und Zugriffs-Token).

Der folgende Code zeigt, wie ein pandas DataFrame mit den Modell-Vorhersagen so in JSON-Format konvertiert wird, das die Daten von unserem Leaderboard-Server akzeptiert werden.

```python
import pandas as pd

dataset_unseen_path = "./production_dataset/robot_energy_unlabeled.csv"

df = pd.read_csv(dataset_unseen_path)
df['Energy [kJ]'] = pd.Series([0] * len(df))  # Initialize the 'Energy [kJ]' column with zeros

predictions_df = df[['Mission ID', 'Energy [kJ]']]  # Create a new DataFrame with only the 'Energy [kJ]' column

predictions_json = predictions_df.to_json(orient='records')  # Convert the DataFrame to JSON format

predictions_json
```

Wir definieren nun eine Funktion, mit der die Vorhersagen auf den Leaderboard-Server hochgeladen werden können.

*Hinweis*: Installiert ggf. zunächst das Paket `requests` in Eure Umgebung.

```python
import requests

def submit_predictions(group_name: str, group_token: str, predictions_json: str) -> tuple[int, str]:
    url = "https://ids-challenge.imi-services.imi.kit.edu/tp02/evaluate"
    payload = {
            "group_name": group_name,
            "predictions_json": predictions_json,
            }

    headers = {"Authorization": f"Bearer {group_token}"}
    response = requests.post(url, json=payload, headers=headers)
    return response.status_code, response.text
```

Gruppennamen und Zugriffstoken müssen **genau wie ausgegeben** eingetragen werden, damit die Ergebnisse vom Server akzeptiert werden!

Gib Gruppennamen und Token in den ersten zwei Codezeilen unten ein, um die Vorhersage-Einreichung zu testen.

Überprüfe die Ausgabe, um zu sehen, ob deine Einreichung erfolgreich war.
Falls ja, dann erscheint der Name eurer Gruppe auch direkt im Leaderboard :)

Deadline für den Upload von Ergebnissen ins Leaderboard ist am Sonntag, 26.07.2026.

```python
group_name = "YourGroupName"
group_token = "YourGroupToken"

response_code, response_body = submit_predictions(group_name, group_token, predictions_json)

print(f"Response code: {response_code}")

response_body
```
