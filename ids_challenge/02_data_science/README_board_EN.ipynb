To participate in the leaderboard, all you need to do is to upload your predictions on the unseen dataset to our server, providing your group name and access token.

The following code demonstrates how to convert your pandas DataFrame with your model's predictions to JSON format which is accepted by our leaderboard server.

```python
import pandas as pd

dataset_unseen_path = "./production_dataset/robot_energy_predictions.csv"

df = pd.read_csv(dataset_unseen_path)
df['Energy [kJ]'] = pd.Series([0] * len(df))  # Initialize the 'Energy [kJ]' column with zeros

predictions_df = df[['Mission ID', 'Energy [kJ]']]  # Create a new DataFrame with only the 'Energy [kJ]' column

predictions_json = predictions_df.to_json(orient='records')  # Convert the DataFrame to JSON format

predictions_json
```

We now define a function which can be used to upload your predictions to the leaberboard server.

*Hint:* You might have to install the paket `requests` to your environment first.

```python
import requests

def submit_predictions(group_name: str, group_token: str, predictions_json: str) -> tuple[int, str]:
    url = "https://ids-challenge.imi-services.imi.kit.edu/tp02/evaluate"
    payload = {
            "group_name": group_name,
            "predictions_json": predictions_json,
            }

    headers = {"Authorization": f"Bearer {group_token}"}
    response = requests.post(url, json=payload, headers=headers)
    return response.status_code, response.text
```

You need to provide your group name and group token **exactly as provided** to submit your predictions in JSON format.
